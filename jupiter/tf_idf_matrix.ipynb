```python
import numpy as np

def tokenize(text):
    return text.lower().split()

def tf_idf(documents, vocab):
    N = len(documents)
    tf_idf_matrix = []
    print('_'*100)
    c = 0
    i = 0
    p = round(len(documents)/100)
    for doc in documents:
        doc = str(doc)
        tf_vector = [doc.count(word) / len(doc.split()) for word in vocab]
        idf_vector = [(N / sum([doc.count(w)])) if sum([doc.count(w)])!= 0 else 1 for w in vocab]
        tf_idf_vector = [a * b for a, b in zip(tf_vector, idf_vector)]
        tf_idf_matrix.append(tf_idf_vector)
        if i == c*p:
            c += 1
            print('=', end = '')
        i += 1
    print('  100%')
    return np.array(tf_idf_matrix)

import pandas as pd
data = pd.read_csv('Airplane_Crashes_and_Fatalities_Since_1908.csv',delimiter = ',')
df = pd.DataFrame(data)
descriptions = df["Summary"]

vocab = set()
i = 0
for doc in descriptions:
    vocab.update(tokenize(str(doc)))

tf_idf_matrix = tf_idf(descriptions, list(vocab)) # при вычислении идет загрузка
```

```python
def pack(m):
    rez = ''
    print('_'*100)
    c = 0
    i = 0
    p = round(len(m)/100)
    for row in m:
        for el in row:
            rez += str(el) + ' '
        rez = rez[:-1] + '\n'
        if i == c*p:
            c += 1
            print('=', end = '')
        i += 1
    print('  100%')
    return rez
```

```python
def pack_to_file(m, file):
    f = open(file, 'a')
    print('_'*100)
    c = 0
    i = 0
    p = round(len(m)/100)
    for row in m:
        rez = ''
        for el in row:
            rez += str(el) + ' '
        f.write(rez[:-1] + '\n')
        if i == c*p:
            c += 1
            print('=', end = '')
        i += 1
    print('  100%')
    return rez
```

```python
f = open('tf_idf_matrix.txt', 'w')
f.write(pack(tf_idf_matrix)) # тоже есть полоса загрузки
f.close()
```

```python
pack_to_file(tf_idf_matrix, 'tf_idf_matrix.txt')
```
